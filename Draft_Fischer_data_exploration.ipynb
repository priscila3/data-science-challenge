```python
#cambiar para version final
zf = zipfile.ZipFile('./dataset.zip') 
df_tickets = pd.read_csv(zf.open('tickets.csv'),index_col=0)
df_tasks = pd.read_csv(zf.open('tasks.csv'),index_col=0)
df_editors = pd.read_csv(zf.open('editors.csv'),index_col=0)
df_clients = pd.read_csv(zf.open('clients.csv'),index_col=0)
```

```python
#ver repetidos de ticket_id-sequence number
keep_tasks_col=['sequence_number','ticket_id']
duplicates_tasks=df_tasks[keep_tasks_col]

print("There are "+str(len(duplicates_tasks[duplicates_tasks.duplicated()]))+" duplicates ticket_id-sequence_number")
```

```python
#verificamos match ticket id - task id
print("There are "+ str(df_tickets['id'][~df_tickets['id'].isin(df_tasks['ticket_id'])].dropna().count()) +" tickets without taks")
print("There are "+str(df_tasks['ticket_id'][~df_tasks['ticket_id'].isin(df_tickets['id'])].dropna().count())+" tasks without a ticket")


print("Mean number of tasks per ticket is "+str(df_tasks[['ticket_id','sequence_number']].groupby(['ticket_id']).mean().mean()))


#aux tasks for wordcount difference
df_tasks_agg=df_tasks[['ticket_id','number_words']].groupby(['ticket_id']).sum()

keep_col_tickets=['id','number_words']
df_tickets_aux=df_tickets[keep_col_tickets]
df_tickets_aux=df_tickets_aux.rename(columns={"number_words": "number_words_tickets"})


df_check_words= df_tickets_aux.merge(df_tasks_agg,how='inner',left_on='id',right_on='ticket_id')
df_check_words['word_difference']=df_check_words['number_words']-df_check_words['number_words_tickets']

#rev.abrir por rubro
print("There are "+str(df_check_words.word_difference[df_check_words.word_difference != 0].count())+" tickets with words count difference")   
```
